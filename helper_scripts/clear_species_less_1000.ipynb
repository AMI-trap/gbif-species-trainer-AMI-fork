Clear those species images that have less than 1000 images. Re-download for those, to make sure they are less than 1000 on GBIF instead of because we had a glitch in downloading.

```python
import os
import pandas as pd
import shutil
```

```python
home_dir = "/bask/projects/v/vjgo8416-amber/projects/gbif-species-trainer-AMI-fork/"
data_dir = "/bask/projects/v/vjgo8416-amber/data/gbif-species-trainer-AMI-fork/"
```

```python
less_1000 = pd.read_csv(os.path.join(home_dir,"data_download","species_list_less_1000.csv"))
```

```python
for idx, iRow in less_1000.iterrows():
        
    if iRow["gbif_species_name"] != "NotAvail":
        
        # Get the folder name
        dirname = os.path.join(data_dir,"gbif_images","try_wrapper",iRow["family_name"],iRow["genus_name"],iRow["gbif_species_name"])
        
        if os.path.isdir(dirname):
            print('Deleting...')
            shutil.rmtree(dirname)
            print('Deleted!')
        else:
            print('Not a dir')
```
