```python
import os
import pandas as pd
```

```python
home_dir = "/bask/projects/v/vjgo8416-amber/projects/gbif-species-trainer-AMI-fork/"
data_dir = "/bask/projects/v/vjgo8416-amber/data/gbif-species-trainer-AMI-fork/"
```

```python
data_stats   = pd.read_csv(os.path.join(data_dir,"gbif_images","try_wrapper","data_statistics.csv"))
species_list = pd.read_csv(os.path.join(home_dir,"data_download","uksi-macro-moths-keys.csv"))
```

```python
# Remove any NotAvail rows
species_list_nodup = species_list[species_list["gbif_species_name"] != "NotAvail"]
species_list_nodup.shape
```

```python
# Remove any duplicate rows
species_list_nodup = species_list_nodup.drop_duplicates(subset="gbif_species_name",
                                                       ignore_index=True)
species_list_nodup.shape
```

```python
pd.options.display.max_rows = None
# pd.reset_option("display.max_rows")

species_image_count = pd.merge(species_list_nodup,data_stats[["gbif_species_name","image_count"]],
                               on="gbif_species_name",
                               how="left")
species_image_count
```

```python
# Change NaN to 0
species_image_count.loc[pd.isna(species_image_count["image_count"]),"image_count"] = 0
```

```python
species_image_count
```

```python
# Get a list with <1000 images
species_image_count_less_1000 = species_image_count[species_image_count["image_count"] < 1000]
species_image_count_less_1000.shape
```

```python
# Save this list
species_image_count_less_1000.to_csv(os.path.join(home_dir,"data_download","species_list_less_1000.csv"),index=False)
```
